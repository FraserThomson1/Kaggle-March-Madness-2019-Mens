```python
import numpy as np
import pandas as pd
```

```python
pred1 = pd.read_csv('LR1(clipped).csv')
pred2 = pd.read_csv('NM1.csv')
predFinal = pred2
```

```python
#combine the predictions of two models by taking average
for i in range(len(pred1['Pred'])):
    P1 = pred1['Pred'][i]
    P2 = pred2['Pred'][i]
    PF = P1*0.5+P2*0.5
    predFinal['Pred'][i] = PF
```

```python
predFinal.to_csv('CombinedProba_0.5_0.5',index=False)
```
